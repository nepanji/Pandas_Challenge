# Overview

The purpose of this assignment is to perform an analyis of the district-wide standardized test results for the Py City Schools System. The Py City School Report tracks several factors that impact student achievement in Py City. The report card is designed to provide information to state officials, school officials, parents, students, and any other interested parties with school performance analysis that offers insight to factors that could influence ways to improve education for Py City Schools.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
schoolfile = "Resources/schools_complete.csv"
studentfile = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
schools= pd.read_csv(schoolfile)
students = pd.read_csv(studentfile)

# Combine the data into a single dataset.  
all_school_data = pd.merge(students, schools, how="left", on=["school_name", "school_name"])
```

```python
# Calculate total number of schools
all_school_names = all_school_data['school_name'].unique()
total_schools = len(all_school_names)
```

```python
# Calculate total number of students
total_students = students['student_name'].count()
```

```python
#Calculate the total budget
total_budget = schools['budget'].sum()
```

```python
# Calculate math average
avg_math_score = students['math_score'].mean()
```

```python
# Calculate reading average
avg_reading_score = students['reading_score'].mean()
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
passing_math_score = all_school_data.loc[all_school_data['math_score'] >=70]
students_passing_math = passing_math_score['student_name'].count()
passing_math_percent = (students_passing_math / total_students) * 100
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading_score = all_school_data.loc[all_school_data['reading_score'] >=70]
students_passing_reading = passing_reading_score['student_name'].count()
passing_reading_percent = (students_passing_reading / total_students) * 100
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
passing_overall_score = all_school_data.loc[(all_school_data['reading_score'] >=70) & (all_school_data['math_score'] >=70)]
students_passing_overall = passing_overall_score['student_name'].count()
passing_overall_percent = (students_passing_overall / total_students) * 100
```

```python
# Calculate percent passing math by school name
schools_passing_math = all_school_data.loc[all_school_data['math_score'] >=70].groupby('school_name').count()['size']
schools_passing_math_percent = pd.DataFrame((schools_passing_math / total_students) * 100).rename({
    'size': "% Passing Math"
}, axis = 1)
schools_passing_math_percent.head()
```

```python
# Calculate percent passing reading by school name
schools_passing_reading = all_school_data.loc[all_school_data['reading_score'] >=70].groupby('school_name').count()['size']
schools_passing_reading_percent = pd.DataFrame((schools_passing_reading / total_students) * 100).rename({
    'size': "% Passing Reading"
}, axis = 1)
schools_passing_reading_percent.head()
```

```python
# Calculate percent overall passing by school name
schools_passing_overall = all_school_data.loc[(all_school_data['math_score'] >=70) & (all_school_data['reading_score'] >=70)].groupby('school_name').count()['size']
schools_passing_overall_percent = pd.DataFrame((schools_passing_overall / total_students) * 100).rename({
    "size": "% Passing Overall"
}, axis = 1)
schools_passing_overall_percent.head()
```

## Scores by School Spending

* This table breaks down school performances based on the following averages vs Spending Ranges (Per Student):
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate

```python
# Find the average of the % passing
pass_percent = (schools_passing_math_percent['% Passing Math'] + schools_passing_reading_percent['% Passing Reading'])/ 2
```

```python
spending_scores = pd.DataFrame({
    'Total Students': total_students,
    'Per Student Budget': budget_per_student['Per Student Budget'],
    'Average Math Score': math_avg_school,
    'Average Reading Score': read_avg_school,
    '% Passing Math': schools_passing_math_percent['% Passing Math'],
    '% Passing Reading': schools_passing_reading_percent['% Passing Reading'],
    'Overall Passing Rate': pass_percent
})
spending_by_scores_df = (spending_scores)
```

```python
# Create the bins for the data
bins = [575, 584.9, 629.9, 644.9, 680.0]

# Create bin names
bin_names = ['Spending < $585', '$585 - $630', '$630 - $645', '$645 - $680']
```

```python
# Place the data series into a new column in the DataFrame
spending_by_scores_df['Spending Ranges (Per Student)'] = pd.cut(spending_by_scores_df['Per Student Budget'], bins, labels=bin_names)

```

```python
# Create a Groupby based on new column
spending_group = spending_by_scores_df.groupby('Spending Ranges (Per Student)')
```

```python
spending_group[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate']].mean()
```
